```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import sklearn.preprocessing import OneHotEncoder

def sigmoid(z):
    return 1/(1+ np.exp(-z))

plt.plot(np.arange(-5,5,0.1), sigmoid(np.arange(-5,5,0.1)))
plt.title("Visualization of sigmoid function")
plt.show()
```
